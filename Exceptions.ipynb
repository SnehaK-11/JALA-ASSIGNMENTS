```python
## Write a program to generate Arithmetic Exception without exception handling
num1 = 10
num2 = 0
result = num1 / num2
print("Result:", result) 
```

```python
## Handle the Arithmetic exception using try-catch block
try:
    result = 10 / 0
except ZeroDivisionError:
    print("Arithmetic Exception: Division by zero")
```

```python
## Write a method which throws exception, Call that method in main class without try block
def divide(a, b):
    return a / b
result = divide(10, 0)
```

```python
## Write a program with multiple catch blocks
try:
    num = int("abc")    
    result = 10 / 0    
except ZeroDivisionError:
    print("Arithmetic Exception: Division by zero")
except ValueError:
    print("Value Error: Invalid conversion to int")
except Exception as e:
    print("Some other exception:", e)
```

```python
## Write a program to throw exception with your own message
try:
    raise ValueError("This is my Jupyter Notebook")
except ValueError as e:
    print("Caught an exception:", e)
```

```python
## Write a program to create your own exception
class MyException(Exception):
    pass
try:
    raise MyException("This is a my own exception")
except MyException as e:
    print("Caught exception:", e)
```

```python
## Write a program with finally block
try:
    result = 10 // 2
    print("Result:", result)
except ZeroDivisionError:
    print("Division by zero occurred")
finally:
    print("This block always executes")
```

```python
## Write a program to generate IOException
file_path = "file.txt" 
try:
    with open(file_path, "w") as file:
        file.write("Hello, World!")
except IOError:
    raise IOError("IOException: Unable to write to the file")
print("File written successfully")
```
