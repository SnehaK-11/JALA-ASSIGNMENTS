```python
## Write a function for arithmetic operators(+,-,*,/)
def arithmetic_operations(a,b):
    print("Addition is:",a + b)
    print("Subtraction is:", a - b)
    print("Multiplication is:",a * b)
    print("Division is:",a / b)
arithmetic_operations(10, 5)
```

```python
## Write a method for increment and decrement operators(++, --)
def incre_decre(n):
    n += 1
    print("After Increment:", n)
    n -= 1
    print("After Decrement:", n)
incre_decre(20)
```

```python
## Write a program to find the two numbers equal or not.
x = 20
y = 10
if x == y:
    print("Both are equal")
else:
    print("Both are not equal")
```

```python
## Program for relational operators (<,<==, >, >==)
a = 50
b = 30
print("a < b is:", a < b)
print("a <= b is:", a <= b)
print("a > b is:", a > b)
print("a >= b is:", a >= b)
```

```python
## Print the smaller and larger number
a = 100
b = 30
if a > b:
    print("Larger:", a, "Smaller:", b)
else:
    print("Larger:", b, "Smaller:", a)
```
