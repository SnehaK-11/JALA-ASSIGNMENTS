```python
## Create a class with PRIVATE fields, private method and a main method. Print the fields in main method. Call the private method in main method.Create a sub class and try to access the private fields and methods from sub class
class Parent:
    def __init__(self):
        self.__private_field = "I am a private field"
    def __private_method(self):
        print("I am a private method")
    @staticmethod
    def main():
        obj = Parent()
        print(obj.__private_field)
        obj.__private_method()
class Child(Parent):
    def try_access(self):
        print("Cannot access private members of Parent class")
Parent.main()
child = Child()
child.try_access()
```

```python
## Create a class with PROTECTED fields and methods. Access these fields and methods from any other class in the same package. Also, Access the PROTECTED fields and methods from child class located in a different packageAccess the PROTECTED fields and methods from any class in different package
# File: package1/module1.py
class Parent:
    def __init__(self):
        self._protected_field = "I am a protected field"
    
    def _protected_method(self):
        return "I am a protected method"
# Access within the same module/package
class SamePackageClass:
    def access_parent(self):
        p = Parent()
        print(p._protected_field)        
        print(p._protected_method()) 
# File: package2/module2.py
class Child(Parent):  
    def access_parent_protected(self):
        print(self._protected_field)      # Access protected field from parent
        print(self._protected_method())   # Access protected method from parent

# ------------------------
# File: package3/module3.py
class DifferentPackageClass:
    def access_parent(self):
        p = Parent()
        print(p._protected_field)         # Access protected field
        print(p._protected_method())      # Access protected method
# Main script 
if __name__ == "__main__":
    print("Accessing from same package:")
    spc = SamePackageClass()
    spc.access_parent()

    print("\nAccessing from child in different package:")
    child = Child()
    child.access_parent_protected()

    print("\nAccessing from different package class:")
    dpc = DifferentPackageClass()
    dpc.access_parent()
```

```python
## Create a class with PUBLIC fields and methods.Access the public methods and fields from any class in the same package or different package
# File: package1/module1.py
class MyClass:
    def __init__(self):
        self.public_field = "I am a public field"

    def public_method(self):
        return "I am a public method"

# Access within the same package
class SamePackageClass:
    def access_myclass(self):
        obj = MyClass()
        print(obj.public_field)       # Access public field
        print(obj.public_method())    # Access public method

# ------------------------
# File: package2/module2.py

class DifferentPackageClass:
    def access_myclass(self):
        obj = MyClass()
        print(obj.public_field)       # Access public field
        print(obj.public_method())    # Access public method

# File: package3/module3.py
class ChildClass(MyClass):
    def access_parent(self):
        print(self.public_field)      # Access public field
        print(self.public_method())   # Access public method
# Main script
if __name__ == "__main__":
    print("Accessing from same package:")
    spc = SamePackageClass()
    spc.access_myclass()

    print("\nAccessing from different package:")
    dpc = DifferentPackageClass()
    dpc.access_myclass()

    print("\nAccessing from child class in different package:")
    child = ChildClass()
    child.access_parent()
```
