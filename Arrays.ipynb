```python
## Write a function to add integer values of an array
numbers = [10, 23,43,64, 45]
total = sum(numbers)
print("Sum of array:", total)
```

```python
## Write a function to calculate the average value of an array of integers
def avg_array(arr):
    total = sum(arr)      
    count = len(arr)     
    return total // count  
numbers = [100, 200, 300, 400, 500]
print("Average of array:", avg_array(numbers))
```

```python
## Write a program to find the index of an array element
arr = [110, 230, 340, 420, 510]
element = 340
print("Index of", element, "is", arr.index(element))
```

```python
## Write a function to test if array contains a specific value
def contains(arr, value):
    return value in arr
numbers = [0,1, 2, 3, 4, 5]
print(contains(numbers, 0))  
print(contains(numbers, 8)) 
```

```python
## Write a function to remove a specific element from an array
def remove_element(arr, value):
    if value in arr:
        arr.remove(value)
    return arr
numbers = [0,1, 2, 3, 4, 5]
print(remove_element(numbers, 3))  
```

```python
## Write a function to copy an array to another array
def copy_array(arr):
    return arr.copy()
original = [1, 2, 3]
copy = copy_array(original)
print(copy)
```

```python
## Write a function to insert an element at a specific position in the array
def insert_element(arr, index, value):
    arr.insert(index, value)
    return arr
numbers = [11, 20, 42, 35]
print(insert_element(numbers, 2, 23)) 
```

```python
## Write a function to find the minimum and maximum value of an array
def min_max(arr):
    return min(arr), max(arr)
numbers = [102, -5, 320, 20]
print(min_max(numbers)) 
```

```python
## Write a function to reverse an array of integer values
def reverse_array(arr):
    return arr[::-1]
numbs = [1, 2, 3, 4]
print(reverse_array(numbs))
```

```python
## Write a function to find the duplicate values of an array
def duplicates(arr):
    return list(set([x for x in arr if arr.count(x) > 1]))
numbers = [1, 2,3,3,3,4,9,1,2,2, 3, 4, 4, 5]
print(duplicates(numbers)) 
```

```python
## Write a program to find the common values between two arrays
def common_elements(arr1, arr2):
    return list(set(arr1) & set(arr2))
arr1 = [1, 2, 3, 4,5,6]
arr2 = [3, 4, 5, 6,7,8]
print(common_elements(arr1, arr2))
```

```python
## Write a method to remove duplicate elements from an array
def remove_duplicates(arr):
    return list(set(arr))
numbers = [1,2,2,3,4,5,6,4,3,24]
print(remove_duplicates(numbers))
```

```python
## Write a method to find the second largest number in an array
def second_largest(arr):
    unique = list(set(arr)) 
    unique.sort()            
    return unique[-2]        
numbers = [10,20,30,400,80,30]
print("Second largest number is:", second_largest(numbers))
```

```python
## Write a method to find number of even number and odd numbers in an array
def count_even_odd(arr):
    even = sum(1 for num in arr if num % 2 == 0)
    odd = len(arr) - even
    return even, odd
numbers = [100,40,56,34,89,95,86]
even_count, odd_count = count_even_odd(numbers)
print("Even numbers:", even_count)
print("Odd numbers:", odd_count)
```

```python
## Write a function to get the difference of largest and smallest value
def difference_max_min(arr):
    return max(arr) - min(arr)
numbers = [101, 230, 25, 3230, 1532]
print("Difference is:", difference_max_min(numbers))
```

```python
## Write a method to verify if the array contains two specified elements(12,23)
def contains_two_elements(arr, val1, val2):
    return val1 in arr and val2 in arr
numbers = [5, 12, 23,90,7, 10]
print(contains_two_elements(numbers, 12, 23)) 
numbers2 = [12, 5, 7]
print(contains_two_elements(numbers2, 12, 23)) 
```
