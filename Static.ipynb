```python
## Define a static variable and access it through a class
class Student:
    Name = "Sneha" 
print(Student.Name)
```

```python
## Define a static variable and access it through an instance
class Student:
   Name = "Sneha"
s1 = Student()
print(s1.Name)
```

```python
## Define a static variable and change it within the instance
class Student:
   Name = "Sneha"
s1 = Student()
s1.Name = "Keerthi Kona"   
print(s1.Name)      
print(Student.Name) 
```

```python
## Define a static variable and change it within the class
class Student:
   Name = "Sneha"
Student.Name = "Keerthi"
print(Student.Name)
```
