```python
## Different ways creating a string
str1 = "Sneha"    ## double quotes      
str2 = 'Keerthi'      ## single quotes     
str3 = """Kona"""   ## triple qyoutes
print(str1, str2, str3)
```

```python
## Concatenating two strings using + operator
string1 = "Sneha"
string2 = "Keerthi"
result = string1 + " " + string2
print(result)
```

```python
## Finding the length of the string
Name = "Sneha Keerthi Kona"
print(len(Name))
```

```python
## Extract a string using Substring
Name = "Sneha Keerthi Kona"
print(Name[0:4])  
print(Name[3:]) 
```

```python
## Searching in strings using index()
Name = "Sneha Keerthi Kona"
print(Name.index("Kona"))
```

```python
## Matching a String Against a Regular Expression
import re
Name = "Sneha1111"
pattern = r"Keerthi\d+"  
if re.fullmatch(pattern, Name):
    print("Matches")
else:
    print("Does not match")
```

```python
## Comparing strings
string1 = "Sneha"
string2 = "Keerthi"
print(string1 == string2)   
print(string1 != string2)    
```

```python
## startsWith(), endsWith() and compareTo()
Name = "Sneha Keerthi Kona"
print(Name.startswith("Sneha"))
print(Name.endswith("Keerthi"))  
print((Name > "Sneha"))        
```

```python
## Trimming strings with strip()
Name = "  Sneha Keerthi Kona  "
print(Name.strip())
```

```python
## Replacing characters in strings with replace()
Name = "Sneha Keerthi Kona"
print(Name.replace("Kona", "K"))
```

```python
## Splitting strings with split()
Name = "Sneha Keerthi Kona"
words = Name.split()  
print(words)         
```

```python
## Converting integer objects to Strings
num = 100
str_num = str(num)
print(str_num, type(str_num))
```

```python
## Converting to uppercase and lowercase
Name = "Sneha Keerthi Kona"
print(Name.upper()) 
print(Name.lower()) 
```
