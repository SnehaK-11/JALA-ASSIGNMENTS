```python
## Write a program to find Armstrong number or not
# Take input from the user
number = int(input("Enter a number: "))
sum = 0
temp = number
while temp > 0:
    digit = temp % 10
    sum += digit ** 3
    temp //= 10
if number == sum:
    print(number, "is an Armstrong number")
else:
    print(number, "is not an Armstrong number")
```

```python
## Write a program to find the prime or not.
number = int(input("Enter a number:"))
if number > 1:
    for i in range(2, number):
        if number % i == 0:
            print(number, "is not a prime number")
            break
    else:
        print(number, "is a prime number")
else:
    print(number, "is not a prime number")
```

```python
## Write a program to palindrome or not.
n = input("Enter a number: ")
if n == n[::-1]:
    print(n, "is a Palindrome")
else:
    print(n, "is not a Palindrome")
```

```python
## Program to check whether a number is EVEN or ODD using switch
## Python doesn't have switch keyword, it has match keyword
num = int(input("Enter a number: "))
match num % 2:
    case 0:
        print(num, "is Even")
    case 1:
        print(num, "is Odd")
```

```python
## Print gender (Male/Female) program according to given M/F using switch
gender = input("Enter gender (M/F): ")
match gender:
    case "M":
        print("Male")
    case "F":
        print("Female")
    case _:
        print("Invalid input")
```
