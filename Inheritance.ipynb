```python
"""
A, B and C are classes
A is a super class. B is a sub class of A. C is a sub class of B. 
Create three methods in each class, 2 methods are specific to each class and third 
method (override method) should be in all three Classes A, B and C
Create a class with main method. Create an object for each class A, B and C in main 
method and call every method of each class using its own object/instance.
Call an overridden method with super class reference to B and C classâ€™s objects
Runtime Polymorphism with Data Members/Instance variables, Repeat the above 
process only for data members
"""

class A:
    def a_method1(self):
        print("A specific method 1")

    def a_method2(self):
        print("A specific method 2")

    def show(self):   
        print("Show method from class A")
class B(A):
    def b_method1(self):
        print("B specific method 1")

    def b_method2(self):
        print("B specific method 2")

    def show(self):   
        print("Show method from class B")
class C(B):
    def c_method1(self):
        print("C specific method 1")

    def c_method2(self):
        print("C specific method 2")

    def show(self):   
        print("Show method from class C")
class Main:
    def main():
        objA = A()
        objB = B()
        objC = C()
        objA.a_method1()
        objA.a_method2()
        objA.show()
        objB.b_method1()
        objB.b_method2()
        objB.show()
        objC.c_method1()
        objC.c_method2()
        objC.show()
        print("\nCalling overridden method using superclass reference:")
        ref = A()
        ref = B()
        ref.show()  
        ref = C()
        ref.show()  
    main()
```
